# Exercise 1

```python
import numpy as np

import matplotlib.pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='blue', linestyle='--',
             marker='o', markerfacecolor='red',
             ecolor='black', capsize=5,
             label='Data')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

# Exercise 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(0, 50, 400)
# Define the function for y
y = np.exp(-x / 10) * np.sin(x)

# Create the plot
plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x, y)

# Customizing the plot as per the exercise requirements
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')

# Set the y limit
plt.ylim([-1, 1])

# Add grid with 25% opacity
plt.grid(True, alpha=0.25)

# Add the equation of the function as the title
plt.title(r'$y = e^{-\frac{x}{10}} \sin(x)$', fontsize=20)

# Show the plot
plt.show()
```
