### Task 4

```python
import glob
import os

# Function to rename filenames
def rename_my_file(old_file_name):
    # Dictionary mapping month abbreviations to numerical format
    month_mapping = {
        'jan': '01', 'feb': '02', 'mar': '03',
        'apr': '04', 'may': '05', 'jun': '06',
        'jul': '07', 'aug': '08', 'sep': '09',
        'oct': '10', 'nov': '11', 'dec': '12'
    }

    # Splitting the filename into components
    parts = old_file_name.split('_')
    month, day = parts[0].split('-')
    city = parts[1].split('.')[0]

    # Converting month abbreviation to numerical format
    month_numerical = month_mapping[month]

    # Constructing the new filename
    new_filename = f'{month_numerical}-{day}-{city}.txt'

    return new_filename

# Specify the directory path containing the .txt files
directory_path = '/Users/keefe/Documents/GitHub/learning-portfolio-keefefons/files, folders & os  exercises/os-collaboration-exercise-data'

# Change the current directory to the specified directory
os.chdir(directory_path)

# Get a list of all .txt files in the folder
txt_files = glob.glob('*.txt')

# Rename each .txt file
for old_file_name in txt_files:
    new_file_name = rename_my_file(old_file_name)
    # Check if the new filename already exists
    if os.path.exists(new_file_name):
        print(f"Skipping {old_file_name} as {new_file_name} already exists.")
    else:
        os.rename(old_file_name, new_file_name)
        print(f'Renamed {old_file_name} to {new_file_name}')
```

### Task 7

```python
import glob
import shutil
import os

# List of cities
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

# Iterate over each city
for city in cities:
    # List all files related to the current city
    city_files = glob.glob(f'*{city.lower()}.txt')
    
    # Destination folder for the current city
    city_folder = city
    
    # Create the folder for the current city if it doesn't exist
    if not os.path.exists(city_folder):
        os.makedirs(city_folder)
    
    # Move files to the corresponding folder of the current city
    for file in city_files:
        destination = os.path.join(city_folder, os.path.basename(file))
        if not os.path.exists(destination):
            shutil.move(file, city_folder)
            print(f'Moved {file} to {city_folder}')
        else:
            print(f'File {os.path.basename(file)} already exists in {city_folder}')
```
